```python
import os, shutil
PROJECT_DIR = "/content/drive/MyDrive/SAR/private-bdi"

# ic ice git reposu kalintisi var mi kontrol et ve temizle (HarmoniDiff'in kendi .git'i gibi)
for root, dirs, files in os.walk(PROJECT_DIR):
    if root == PROJECT_DIR:
        continue  # ana klasorun kendi .git'i olacak, ona dokunma
    if ".git" in dirs:
        inner_git = os.path.join(root, ".git")
        print(f"Ic ice .git bulundu, siliniyor: {inner_git}")
        shutil.rmtree(inner_git)

gitignore_content = """
data/
outputs/
models/HarmoniDiff/checkpoints/
models/HarmoniDiff/demo/
models/HarmoniDiff/visualization*/
__pycache__/
.ipynb_checkpoints/
*.tif
*.pt
"""
with open(f"{PROJECT_DIR}/.gitignore", "w") as f:
    f.write(gitignore_content)
print("gitignore yazildi")
```
